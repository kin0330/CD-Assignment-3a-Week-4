### Question 1
When was the first COVID case reported in the Netherlands, and how long did it take for more than 10,000 cases to be reported?

```python
NL_filter = 'Netherlands'
```

```python
Netherlands_table = covid_df[covid_df['Country'] == NL_filter]
```

```python
Netherlands_table[(Netherlands_table['Cumulative_cases'] > 0) & (Netherlands_table['Cumulative_cases'] < 10000)]
```

The first COVID case in the Netherlands was reported on March 1, 2020, and it took 28 days for over 10,000 cases to be reported, which occurred on March 29, 2020.

### Question 2
When did Belgium report its first COVID-related death, and how much time passed before more than 1,000 cases were reported?

```python
BE_filter = 'Belgium'
```

```python
Belgium_table = covid_df[covid_df['Country'] == BE_filter]
```

```python
Belgium_table[(Belgium_table['Cumulative_deaths'] > 0) & (Belgium_table['Cumulative_deaths'] < 1000)]
```

Belgium reported its first COVID-related death on March 9, 2020, and it took 21 days until more than 1,000 cases were reported on March 30, 2020.
